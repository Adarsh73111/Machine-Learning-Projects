```python
img = cv2.imread("133813443263992628.jpg")
# print(img)
print(img.shape)
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
X = img.reshape((-1,3))
print(X.shape)
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
k = 5
model = KMeans(n_clusters=k)
model.fit(X)
```

```python
centroids = model.cluster_centers_
```

```python
print(centroids)
```

```python
colors = np.array(centroids,dtype="uint8")
```

```python
print(colors)
```

```python
i = 1
for color in colors:
  # print(color)
  plt.subplot(1,k,i)
  i = i+1
  mat = np.zeros((100,100,3),dtype = 'uint8')
  mat[:,:,:] = color
  plt.imshow(mat)
  plt.axis("off")

plt.show()
```
