```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
```

```python
X = california['data']
```

```python
X.shape
```

```python
mu = X.mean(axis=0)
X_ = (X-mu)
X.shape
```

```python
S = np.dot(X_.T, X_)
S.shape
```

```python
lambdas, vectors = np.linalg.eig(S)
```

```python
lambdas.shape
```

```python
lambdas[:2]
```

```python
vectors.shape
```

```python
V = vectors[:,: 2]
V.shape
```

```python
X_.shape
```

```python
V.shape
```

```python
X_new = np.dot(X_, V)
X_new.shape
```

```python
X_new
```

```python
lambdas
```

```python
total = sum(lambdas)
```

```python
print(total)
```

```python
np.round(lambdas / total, 3)
```

```python
explained_variance = np.round(lambdas / total, 3)
explained_variance
```

```python
print(explained_variance[: 2].sum())
```

```python
explained_variance
```

```python
np.cumsum(explained_variance)
```

```python
cummulative_variance = np.cumsum(explained_variance)
cummulative_variance
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range (1,len(cummulative_variance)+1), cummulative_variance, marker = "o")
plt.xlabel('number of components')
plt.ylabel('explained variance')
plt.grid("on")
plt.show()
```
