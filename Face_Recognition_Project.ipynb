```python
model = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = model.detectMultiScale(newImg,1.3,2)
```

```python
faces.shape
```

```python
faces
```

```python
for f in faces :
  x,y,w,h = f
  cv2.rectangle(newImg, (x,y),(x+w,y+h),(0,255,0),2)

plt.imshow(newImg)
plt.axis("off")
plt.show()
```
